```python
import os


def list_all_files_and_folders(directory):
    for root, dirs, files in os.walk(directory):
        print(f"Root: {root}")
        for dir_name in dirs:
            print(f"Directory: {dir_name}")
        for file_name in files:
            print(f"File: {file_name}")

list_all_files_and_folders(directory)

# Example usage:
# list_all_files_and_folders('/path/to/your/directory')
```

```python
import pandas as pd
import os
from tqdm import tqdm

directory = 'F:/지하수 기초조사 성과/'

def list_files_and_folders_to_df(directory):
    data = {'Root': [], 'Type': [], 'Name': []}
    for root, dirs, files in tqdm(os.walk(directory), desc='Processing files'):
        for dir_name in dirs:
            data['Root'].append(root)
            data['Type'].append('Directory')
            data['Name'].append(dir_name)
        for file_name in files:
            data['Root'].append(root)
            data['Type'].append('File')
            data['Name'].append(file_name)
    return pd.DataFrame(data)

df_files_and_folders = list_files_and_folders_to_df(directory)
print(df_files_and_folders)

df_files_only = df_files_and_folders[df_files_and_folders['Type'] == 'File']
print(df_files_only)
```

```python
df_files_fold = df_files_and_folders[df_files_and_folders['Type'] == 'Directory']
print(df_files_fold)
```

```python
df_files_gims = df_files_fold[df_files_fold['Name'].str.contains('GIMS', na = False)]
print(df_files_gims)
df_files_gims.to_csv('GIMS_list.csv', encoding='euc-kr',index = False)
```

```python
df_files_sungapum = df_files_only[df_files_only['Name'].str.contains('.*성과품*', na=False)]
print(df_files_sungapum)
```
