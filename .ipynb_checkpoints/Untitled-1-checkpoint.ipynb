```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("4번_데이터항목.csv")
df_test = pd.read_csv("4번_데이터항목_전처리.csv")
total_df = pd.DataFrame(columns= ['구분','관측소명','현장관정번호','년도','분기','성분'])
```

```python
for i in range(0,len(df_test)) :
    place_name = df_test.iloc[i,1]
    sort_name = df_test.iloc[i,2]
    element = df_test.iloc[i,5]
    df_filter_1 = df.loc[df['관측소명']==f'{place_name}']
    df_filter_1 = df_filter_1[df_filter_1['현장관정번호'] == f'{sort_name}']
    plus_df = df_filter_1[['구분', '관측소명','현장관정번호','년도','분기',f'{element}']]
    plus_df['성분명'] = element
    plus_df = plus_df.rename(columns={f'{element}':'성분'})
    plus_df = plus_df.sort_values(by = '년도')
    total_df = pd.concat([total_df,plus_df])
```
